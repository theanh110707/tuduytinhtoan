### W5A1 
### Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a,b) -> int:

```python
def so_lon_hon(a,b):
    if a>b:
        return a
    elif b>a:
        return b
    else:
        return 0
so1=int(input())
so2=int(input())        
so_lon=so_lon_hon(so1, so2)
print(so_lon)                
```

### W5A2 
### Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]

```python
from typing import Tuple
def swap(a:int,b:int) -> Tuple[int,int]:
    a,b=b,a
    return a,b
x,y=map(int,input().split())
x,y=swap(x,y)
print(x,y)
```

### W5A3 
###  hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
import math
def so_nguyen_to(a):
    if a<2:
        return False
    else:
        for i in range(2,int(math.sqrt(a))+1):
            if a% i==0:
                return False
        return True    
n=int(input())
snt=so_nguyen_to(n)
print(snt)            
```

### W5A4 V
### Iết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_hoan_hao(a):
    if a<0:
        return False
    else:
        sum=0
        for i in range(1,a):
            if a % i==0:
                sum +=i
        return sum
n=int(input())
so=so_hoan_hao(n)
print(so)            
```

### W5A5 
### Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của  phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

```python
def vi_tri(lst,k):
    if k in lst:
        return lst.index(k)
    else:
        return -1
lit=list(map(int,input().split()))
n=int(input())
a=vi_tri(lit,n)
print(a)    
```

### W5A6 
### Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python

def giai_thua(n):
    giaithua=1
    for i in range(1,n+1):
        giaithua *=i
    return giaithua    
a=int(input())
k=giai_thua(a)
print(k)
    
```

### Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat \) 
### Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
### Phép trừ \( - \) 
### Phép cộng \( + \) 
###  Phép chia \( / \) 
### Phép nhân \( * \) 
### Kết quả làm tròn đến chữ số thập phân thứ hai. 
### Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
### Đầu ra: Kết quả phép toán.

```python
def tinh_toan(num1,num2):
    tong=num1+num2
    
    hieu=num1-num2
    
    tich=num1*num2
    if num2 != 0:
        thuong = round(num1 / num2, 2)
    else:
        thuong = "Không thể chia cho 0"
    return round(tong,2),round(hieu,2),round(tich,2),round(thuong,2)
a=float(input())
b=float(input())
tong,hieu,tich,thuong=tinh_toan(a,b)
print(tong,hieu,tich,thuong)
    
```

### W5A9 
### Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def tong_chu_so(n):
    tong=0
    for chu_so in str(n):
        tong +=int(chu_so)
    return tong
a=int(input())
b=tong_chu_so(a)
print(b)    
```
