### W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyen x

```python
def amstronhg(a):
                   
    n=len(str(a))
    
    sum=0
    for i in str(a):
        x=int(i)**n
        sum += x
    if sum==a:
        return True
    else:
        return False
n=int(input())
x=amstronhg(n)
print(x)               
```

### W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_doi_xung(n):
    
    if n < 0:
        return False
    
    original = n
    reversed_num = 0
    while n > 0:
        digit = n % 10        
        reversed_num = reversed_num * 10 + digit  
        n = n // 10               
    return reversed_num == original


n = int(input())
if so_doi_xung(n):
    print(f"{n} là số đối xứng.")
else:
    print(f"{n} không phải là số đối xứng.")
        
```

### W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def so_ky_dieu(a):
    so_tam = a
    while so_tam >= 10:
        tong = 0
        for i in str(so_tam): 
            tong += int(i)
        so_tam = tong        
    
    return so_tam == 1        

a = int(input("Nhập số: "))
if so_ky_dieu(a):
    print(f"{a} la so ky dieu")
else:
    print(f"{a} khong la so ky dieu")
```

### W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    val = x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2) #cong thuc tinh dien tich tam giac
    #dien tich=0 thi thang hang
    #dien tich !=0 thi khong thang hang
    if val == 0:
        return True
    else:
        return False

```

### Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),(0,3)], p(2,1)

```python
def point_in_quadrilateral(quad, p):
    # quad = [(x1,y1), (x2,y2), (x3,y3), (x4,y4)]
    # p = (px, py)
    
    px, py = p
    signs = []
    
    for i in range(4):
        # Lấy tọa độ 2 đỉnh của một cạnh (A và B)
        # i + 1 dùng % 4 để đỉnh cuối nối lại với đỉnh đầu
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % 4]
        
        # Công thức tích có hướng (Cross Product)
        cp = (x2 - x1) * (py - y1) - (y2 - y1) * (px - x1)
        
        # Nếu cp == 0 thì điểm nằm trên biên cạnh đó
        if cp == 0:
            # Kiểm tra xem điểm p có nằm trong đoạn AB không (tránh nằm ngoài đường thẳng kéo dài)
            if min(x1, x2) <= px <= max(x1, x2) and min(y1, y2) <= py <= max(y1, y2):
                continue # Hợp lệ, tiếp tục xét cạnh tiếp theo
            else:
                return False # Nằm trên đường thẳng nhưng ngoài đoạn thẳng
        
        signs.append(cp > 0)

    # Nếu tất cả các tích có hướng đều cùng dấu (tất cả True hoặc tất cả False)
    # thì điểm nằm hoàn toàn bên trong.
    return all(signs) or not any(signs)

# Ví dụ chạy thử:
quad_test = [(0,0), (4,0), (4,3), (0,3)]
p_test = (2, 2)
print(point_in_quadrilateral(quad_test, p_test)) # Kết quả: True
```

### W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def tinh_goc_ABC(x1, y1, x2, y2, x3, y3):
    # vecto BA
    ba_x, ba_y = x1 - x2, y1 - y2
    #vecto BC
    bc_x, bc_y = x3 - x2, y3 - y2
    
    # Tích vô hướng 
    tich_vo_huong = ba_x * bc_x + ba_y * bc_y
    
    # Độ dài các vectơ 
    ba = math.sqrt(ba_x**2 + ba_y**2)
    bc = math.sqrt(bc_x**2 + bc_y**2)
    
    # Tránh lỗi chia cho 0 nếu có 2 điểm trùng nhau
    if ba == 0 or bc == 0:
        return 0.0
    
    # Tính cosin và giới hạn giá trị trong khoảng [-1, 1] để tránh lỗi làm tròn
    cos_angle = tich_vo_huong / (ba * bc)
    cos_angle = max(-1.0, min(1.0, cos_angle))
    
    # Tính góc theo radian và chuyển sang độ
    angle_rad = math.acos(cos_angle)
    angle_deg = math.degrees(angle_rad)
    
    return angle_deg

print(tinh_goc_ABC(0, 1, 0, 0, 1, 0))
```

### W5A30 Viết hàm tính hệ số nhị thức C(n, k)

```python
def he_so_nhi_thuc(n,k):
    if k==0:
        return 1
    elif n==0:
        return "khong thoa man"
    elif n>k:
        a=n-k
        n_giai_thua=1
        k_giai_thua=1
        a_giai_thua=1
        for x in range(1,k+1):
            k_giai_thua *=x
        for i in range(1,n+1):
            n_giai_thua *=i
        for j in range(1,a+1):
            a_giai_thua *=j
        c=n_giai_thua/(k_giai_thua*a_giai_thua)
        return c
    elif n==k:
        return 1
    else:
        return " khong thoa man"
    
n,k=map(int,input().split())
c=he_so_nhi_thuc(n,k)
print(c)                
```

### W5A33 Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    #chon lai xe
    if car_type=="XanhSM":
        gia_mo_cua=15000
        gia_km=17000
        phi_cho=700
    elif car_type=="Be":
        gia_mo_cua=15500
        gia_km=18000
        phi_cho=800  
    else:
        return("loai xe khong hop le")
    #tinh tien theo km
    if 0<km<=1:
        gia_tien=gia_mo_cua
    else:
        gia_tien=gia_mo_cua + (km-1)*gia_km
    #tien cho
    gia_cho=wait_minutes*phi_cho
    #tong tien
    tong_tien=gia_tien+gia_cho
    if start_hour>=22 or start_hour <5:
        tong_tien *=1.1
    return tong_tien
car_type=input("chon lai xe: ")
start_hour=int(input("chon gio di: "))
km=float(input("nhap quang duong di chuyen: "))
wait_minutes=int(input("thoi gian cho (neu co): "))    
a=smart_taxi_fare(car_type,start_hour,km,wait_minutes)
print(a)              
       
```

### Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhauNếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n<=0:
        return 0
    else:
        sum=0
        for x in range(1,n+1):
            if len(str(x))!=len(set(str(x))):
                sum +=int(x)
        return sum
a=int(input())
n=sum_1_to_n(a)
print(n)            
```
