```python
h = int(input())

for i in range(1, h + 1):
    
    print(" " * (h - i), end="")
    
    
    if i == 1:
        
        print("*")
    elif i == h:
        
        print("*" * (2 * h - 1))
    else:
       
        print("*" + " " * (2 * i - 3) + "*")
```

### Bài 19: 
### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với sốkế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trongdãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
try:
    A = int(input())

    if A <= 0:
        print("nhap lai")
    else:
        f1, f2 = 1, 1
        
        
        if A < 1:
            ans = 0 
        elif A == 1:
            ans = 1
        else:
            ans = 1
         
            while True:
                next_f = f1 + f2
                if next_f > A:
               
                    ans = f2
                    break
                
            
                f1 = f2
                f2 = next_f
                ans = f2 # Cập nhật kết quả hiện tại

        print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {ans}")

except ValueError:
    print("Dữ liệu nhập vào không hợp lệ.")
```

### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.
