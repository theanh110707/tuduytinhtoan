### W6A20
### Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1


```python
def vi_tri(lst,k):
    for i in range(len(lst)):
        if lst[i]==k:
            return i
    return -1
lst=list(map(int,input().split()))
k=int(input())
print(vi_tri(lst,k))    
```

### W6A21
### Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
words = input().split()
d = {}
for i, word in enumerate(words):
    if word not in d:
        d[word] = []
    d[word].append(i)

# Chuyển list các vị trí thành tuple theo yêu cầu
for word in d:
    d[word] = tuple(d[word])
print(d)
```

### W6A23 VIết hàm xóa đi các số <x có trong mảng

### W6A24
### Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9


```python
def chinh_list(lst):
    arr_1=[]
    arr_2=[]
    for i in lst:
        if i==0:
            arr_1.append(i)
        else:
            arr_2.append(i)
    return arr_1+arr_2
lst=list(map(int,input().split()))
print(chinh_list(lst))


            
```

### W6A27
### Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1


```python
def so_nho_hon(lst,x):
    vi_tri=[]
    for i in range(len(lst)):
        if lst[i]<x:
            vi_tri.append(i)
    if len(vi_tri)>0:
        return vi_tri   
    else:
        return -1
lst=list(map(int,input().split()))
x=int(input())
print(so_nho_hon(lst,x))
    
```

### W6A28
### Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thuc su, hay vo danh(khong tang,khong giam) 

```python
def day(lst):
    n=len(lst)
    if n<2:
        return False
    tang=True
    giam=True
    for i in range(0,n-1):
        if not (lst[i]<lst[i+1]):
            tang=False
        if not (lst[i]>lst[i+1]):
            giam=False
    if tang:
        return "Day tang"
    elif giam:
        return "Day giam" 
    else:
        return "Day vo danh"
lst=list(map(int,input().split()))
print(day(lst))               
```
