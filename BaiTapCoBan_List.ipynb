```python
def list_number(lst):
    if not lst:
        return None
    res = {}
    for i in lst:
        if i not in res:
            res[i] = 1
        else:
            res[i] += 1
            
    res = dict(sorted(res.items()))
    
    return res
lst=list(map(int,input().split()))
print(list_number(lst))
```
